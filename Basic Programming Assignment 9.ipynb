# Q1: Python program to check if the given number is a Disarium Number

```python
n = input('Input a number to check whether it is Disarium or not.')
s = 0
for i in range(len(n)):
    s += (int(n[i]))**(i+1)
    
if s == int(n):
    print('\nNumber {} is Disarium.'.format(n))
else:
    print('\nNumber {} is not Disarium.'.format(n))
```

# Q2: Python program to print all disarium numbers between 1 to 100

```python
for n in range(100):
    s = 0
    n = str(n)
    for i in range(len(n)):
        s += (int(n[i]))**(i+1)
        
    if s == int(n):
        print('\nNumber {} is Disarium.'.format(n))
    else:
        print('\nNumber {} is not Disarium.'.format(n))
```

# Q3: Python program to check if the given number is Happy Number

```python
num = input('Input a number to check whether it is Happy Number or not.')
n = num
while(True):
    s = 0
    n = str(n)
    for i in range(len(n)):
        s += (int(n[i]))**2
    
    if s == 1:
        print('\nNumber {} is Happy Number.'.format(num))
        break   
    elif s == 4:
        print('\nNumber {} is not Happy Number.'.format(num))
        break
    else:
        n = s
```

# Q4: Python program to print all happy numbers between 1 and 100

```python
for num in range(1, 101):
    n = num
    while(True):
        s = 0
        n = str(n)
        for i in range(len(n)):
            s += (int(n[i]))**2
        if s == 1:
            print('\nNumber {} is Happy Number.'.format(num))
            break   
        elif s == 4:
            print('\nNumber {} is not Happy Number.'.format(num))
            break
        else:
            n = s
```

# Q5: Python program to determine whether the given number is a Harshad Number

```python
n = input('Input a number to check whether it is Harshad Number or not.')
s = 0
for i in range(len(n)):
    s += (int(n[i]))
    
if int(n)%s == 0:
    print('\nNumber {} is Harshad Number.'.format(n))
else:
    print('\nNumber {} is not Harshad Number.'.format(n))
```

# Q6: Python program to print all pronic numbers between 1 and 100

```python
for i in range(1, 101):
    
    if int(i**0.5) * (int(i**0.5) + 1) == i:
        print('\nNumber {} is Pronic Number.'.format(i))
    else:
        print('\nNumber {} is not Pronic Number.'.format(i))
```
